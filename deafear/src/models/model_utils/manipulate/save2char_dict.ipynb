```python
DATA_FOLDER = "/Users/trHien/Python/DeafEar/deafear/src/models/model_utils/manipulate/data_convert"
```

```python
df = pd.read_csv("modal_data.csv")

df.head()
```

```python
df.info()
```

```python
df = df[df["Word"].str.len() == 1]
df = df.drop_duplicates(subset=['Word'])
```

```python
df
```

```python
import os
import numpy as np
import json

npys_dict = {}
for row in df.itertuples(index=False):  # index=False to skip index in the tuple
    id, word = row.ID, row.Word  # Extract ID and Word from each row
    
    # if id.endswith(("N", "T")):
    #     id = id[:-1] + "B"  # Replace the last character with "B"
    
    try:
        # Load the numpy file, convert to list, and append to npys_list
        frame_data = np.load(os.path.join(DATA_FOLDER, "landmarks_" + id + ".npy"))
        # frame_data = frame_data.astype(np.float16)
        npys_dict[str(word)] = frame_data.tolist()  # Convert np.array to list for JSON
    except FileNotFoundError:
        # Print missing file's ID and associated word
        print("File not found for ID:", id, "Word:", word)

# Write npys_list to JSON
with open("character_dict.json", "w") as f:
    json.dump(npys_dict, f)
```
